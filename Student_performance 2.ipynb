### Feature encoding

```python
data.head()
```

```python
#encoding binary features
data_copy = data.copy()
data_copy['gender'] = data_copy['gender'].apply(lambda x : 1 if x =='male' else 0)
data_copy['lunch'] = data_copy['lunch'].apply(lambda x : 1 if x =='standard' else 0)
data_copy['test preparation course'] = data_copy['test preparation course'].apply(lambda x : 1 if x =='completed' else 0)
```

```python
import category_encoders as ce
```

```python
#encoding non-binary ordinal features
ord_encoder = ce.OrdinalEncoder()
data_bin = ord_encoder.fit_transform(data_copy[['parental level of education']])
data_copy = data_copy.drop(['parental level of education'], axis=1)
data_copy = pd.concat([data_copy, data_bin], axis=1)

data_copy
```

```python
#encoding non-ordinal categorical features
encoder = ce.OneHotEncoder(cols=['race/ethnicity'], use_cat_names=True) 
type_bin = encoder.fit_transform(data_copy['race/ethnicity'])
data_copy = pd.concat([data_copy, type_bin], axis=1)

data_copy.head()
```

```python
data_copy = data_copy.drop(['race/ethnicity', 'math score alt', 'reading score alt', 'writing score alt'], axis=1)
```

# Prediction

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 5))
sns.heatmap(data_copy.corr(), annot=True)
```

So at this point, I don't think we can use a linear regression model since there are no high correlation scores here. We'll try Adaboost. I will not be normalizing data as decision trees can work with messy data.

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

## Math score prediction


```python
X = data_copy.drop(['math score'], axis=1)
y = data_copy['math score']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
dt = DecisionTreeRegressor(
    max_depth=3, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=50 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

```python
#trying a higher max_depth and number of models
dt = DecisionTreeRegressor(
    max_depth=3, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=60 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

```python
dt = DecisionTreeRegressor(
    max_depth=5, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=100 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

I'll try gradient boosting as well to see if we can reduce the error.

```python
gb = GradientBoostingRegressor(
    max_depth=3, 
    n_estimators=10, 
    random_state=42
)

gb.fit(X_train, y_train)

gb_pred  = gb.predict(X_test)

print(f'MSE score {round(mean_squared_error(y_test, gb_pred),2)}')
```

```python
gb = GradientBoostingRegressor(
    max_depth=5, 
    n_estimators=100, 
    random_state=42
)

gb.fit(X_train, y_train)

gb_pred  = gb.predict(X_test)

print(f'MSE score {round(mean_squared_error(y_test, gb_pred),2)}')
```

Gradient boosting's MSE is close but still higher than Adaboost's results. 

## Reading score prediction

Here I'll be using Adaboost with parameters: max_depth=5, n_estimators=100 as it has shown a better result before.

```python
X = data_copy.drop(['reading score'], axis=1)
y = data_copy['reading score']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
dt = DecisionTreeRegressor(
    max_depth=5, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=100 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

## Writing score prediction

```python
X = data_copy.drop(['writing score'], axis=1)
y = data_copy['writing score']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
dt = DecisionTreeRegressor(
    max_depth=5, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=100 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

```python
dt = DecisionTreeRegressor(
    max_depth=7, 
    random_state=42 
)
ada = AdaBoostRegressor(
    base_estimator=dt, 
    random_state=42,
    n_estimators=150 
)

ada.fit(X_train, y_train)
ada_pred  = ada.predict(X_test)
print(f'MSE score {round(mean_squared_error(y_test, ada_pred),2)}')
```

Thus, we can say that Adaboost would be one of the better models to try to predict student performance. We could also try increasing the number of models in Adaboost and the depth of each tree but that could lead to overfitting.
